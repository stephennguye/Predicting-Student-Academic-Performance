# 6. EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
combined_df = pd.read_csv("data/cleaned/students_combined.csv")
```

```python
combined_df.head()
```

## 6.1 Target Variable Distribution (G3)

```python
plt.figure(figsize=(8,5))
sns.histplot(combined_df["G3"], bins=20, kde=True)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade (0-20)")
plt.ylabel("Count")
plt.show()
```

#### Observation: Most students score between 8-15, but there are some who fail (0) and some who excel (18-20).
--> Predicting G3 is non-trivial, that means lots of middle performers, with fewer extremes

## 6.2 Compare grade distribution by subject

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="subject", y="G3", data=combined_df, palette="Set2")
plt.title("Final Grades by Subject")
plt.show()
```

#### Observation: Math tends to have lower and more spread-out grades compared to Portuguese, which is slightly higher on average --> Keep "subject" column is important

## 6.3 Correlations

```python
numeric_cols = combined_df.select_dtypes(include=np.number).columns

plt.figure(figsize=(12,8))
corr = combined_df[numeric_cols].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
print("Top correlations with G3:", corr["G3"].sort_values(ascending=False).head(10))
```

## 6.4 Categorical Distributions

```python
categorical_cols = combined_df.select_dtypes(include="object").columns

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=combined_df, palette="Set3")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=30)
    plt.show()
```

## 6.5 Relationship between Key Features and G3

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="studytime", y="G3", hue="subject", data=combined_df, palette="coolwarm")
plt.title("Studytime vs Final Grade")
plt.show()
```

#### Observation: 
Longer studytime shows a slight positive relationship with grades, but not strong

This explains why correlation (0.16) is weak: studying more helps, but it's not the only factor

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="failures", y="G3", data=combined_df, palette="Pastel1")
plt.title("Impact of Failures on Final Grade")
plt.show()
```

#### Observation: More past failures --> significantly lower final grades.
--> Failures is a strong categorical predictor.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="higher", y="G3", data=combined_df, palette="Pastel2")
plt.title("Higher Education Aspiration vs Final Grade")
plt.show()
```

#### Observation: Students who want higher education generally have better grades --> This is likely because of motivation.
